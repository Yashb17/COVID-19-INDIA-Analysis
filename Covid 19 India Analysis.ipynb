### Importing all neccessary header files

```python
import requests
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
```

### 1) Data Collection (Making Connection with the website and fetching data)
________

```python
response=requests.get(url="https://www.mohfw.gov.in/") #create connection
data=response.text ##returns all the data of webpage
#print(data)
```

```python
df_list=pd.read_html(data)
```

```python
len(df_list)
```

### Having a look at our dataframe

```python
coronadata=df_list[0]
coronadata=pd.DataFrame(coronadata)
coronadata.columns=["S.No","State","Active Cases","Discharged","Deaths","Total_Cases"]
coronadata
```

________

### 2) Data Preprocessing:
____

```python
coronadata.drop(columns=["S.No"],axis=1,inplace=True)
coronadata
```

______

```python
coronadata.drop([36,37,38,39,40],axis=0,inplace=True)
coronadata.replace(np.nan,0,inplace=True)
coronadata
```

```python
coronadata.dtypes
```

```python
coronadata=coronadata.astype({"State":str,"Active Cases":int,"Total_Cases":int,"Discharged":int,"Deaths":int})
coronadata.dtypes
```

___________

### 3) Exploratory Data Analysis:
___

### Adding extra columns for more precise Analysis

```python
coronadata["Death_Ratio"]=coronadata["Deaths"]/coronadata["Total_Cases"]*100
coronadata["Discharge_Ratio"]=coronadata["Discharged"]/coronadata["Total_Cases"]*100
coronadata
```

```python
print("Total Cases in India: ",coronadata["Total_Cases"].sum())
print("Current Active Cases in India :",coronadata["Active Cases"].sum())
print("Total Discharged Patients in India :",coronadata["Discharged"].sum())
print("Total Deaths in India: ",coronadata["Deaths"].sum())
print("Death Ratio in India: ",coronadata["Deaths"].sum()/coronadata["Total_Cases"].sum()*100)
print("Discharge Ratio or Recovery % in India: ",coronadata["Discharged"].sum()/coronadata["Total_Cases"].sum()*100)
```

```python
india_columns={"Total Cases in India":coronadata["Total_Cases"].sum(),
               "Current Active Cases in India":coronadata["Total_Cases"].sum()-(coronadata["Discharged"].sum()+coronadata["Deaths"].sum()),
               "Total Discharge in India":coronadata["Discharged"].sum(),
               "Total Deaths in India":coronadata["Deaths"].sum(),
               "Death Ratio in India":coronadata["Deaths"].sum()/coronadata["Total_Cases"].sum()*100,
               "Recovery % in India":coronadata["Discharged"].sum()/coronadata["Total_Cases"].sum()*100}
india_data=pd.DataFrame(india_columns,index=[0])
india_data
```

```python
india_columns1={
               "Stats":[
               coronadata["Total_Cases"].sum(),
               coronadata["Total_Cases"].sum()-(coronadata["Discharged"].sum()+coronadata["Deaths"].sum()), 
               coronadata["Discharged"].sum(),
               coronadata["Deaths"].sum(),
               coronadata["Deaths"].sum()/coronadata["Total_Cases"].sum()*100,
               coronadata["Discharged"].sum()/coronadata["Total_Cases"].sum()*100]
               }
india_data1=pd.DataFrame(india_columns1,index=["Total Cases in India",
                                               "Current Active Cases in India",
                                               "Total Discharge in India",
                                               "Total Deaths in India",
                                               "Death Ratio in India",
                                               "Recovery % in India"])
india_data1
```

```python
highest_cases=coronadata.sort_values("Total_Cases",ascending=False)
```

```python
most_deaths=coronadata.sort_values("Deaths",ascending=False)
```

```python
top_discharge_ratios=coronadata.sort_values("Discharge_Ratio",ascending=False)
```

```python
highest_death_ratios=coronadata.sort_values("Death_Ratio",ascending=False)
```

______

### 4) Data Visualization:
____

```python
plt.figure(figsize=(20,8))
sns.barplot(x="State",y="Total_Cases",data=highest_cases.head(10))
plt.title("Top 10 Affected States",fontweight="bold")
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x="State",y="Discharge_Ratio",data=top_discharge_ratios.head(10))
plt.title("Top 10 States with best Recovery Rate",fontweight="bold")
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x="State",y="Death_Ratio",data=highest_death_ratios.head(10))
plt.title("Top 10 States as per Death Ratio",fontweight="bold")
```

```python
sizes=[coronadata["Deaths"].sum(),
       coronadata["Discharged"].sum(),
       coronadata["Total_Cases"].sum()-(coronadata["Discharged"].sum()+coronadata["Deaths"].sum())]
labels=["Deaths","Discharged","Current Active Cases"]
explode=[0.1,0.0,0.0]
fig,ax1 = plt.subplots(figsize=(8,8))
ax1.pie(sizes,labels=labels,explode=explode, autopct='%1.1f%%',shadow=True, startangle=90)
plt.title("Current Situation in India",fontweight="bold")
```
